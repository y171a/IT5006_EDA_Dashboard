```python
# ------------------------------------------------
# 2️⃣ Z-Score Method (robust global filter)
# ------------------------------------------------
# Restrict to realistic daily return ranges
z = np.abs((df['Return'] - df['Return'].mean()) / df['Return'].std())
df_z = df[z < 4].copy()  # keep values within ±4σ (typical for financial data)
print(f"Removed {len(df) - len(df_z)} extreme-return rows via z-score filter")

# ------------------------------------------------
# 3️⃣ IQR Method (per stock) – handles atickmetric volatility
# ------------------------------------------------
def iqr_filter(x, col='Return', k=3.0):
    q1, q3 = x[col].quantile([0.25, 0.75])
    iqr = q3 - q1
    lower, upper = q1 - k * iqr, q3 + k * iqr
    return x[(x[col] >= lower) & (x[col] <= upper)]

df_iqr = (
    df_z.groupby('Name', group_keys=False)
         .apply(iqr_filter, col='Return', k=3.0)
         .reset_index(drop=True)
)
print(f"After IQR filtering: {df_iqr.shape}")
```

```python
# Remove impossible values (negative price/volume)
df_iqr = df_iqr[(df_iqr['close'] > 0) & (df_iqr['volume'] > 0)]

# Limit daily return to ±40% to remove potential split-adjustment artefacts
df_iqr = df_iqr[(df_iqr['Return'] > -0.4) & (df_iqr['Return'] < 0.4)]

print("After sanity filtering:", df_iqr.shape)

# ------------------------------------------------
# 5️⃣ Visual confirmation
# ------------------------------------------------
sns.histplot(df_iqr['Return'], bins=200, kde=True, color='teal')
plt.title("Distribution of Daily Returns (Post-Outlier Removal)")
plt.show()

# Compare original vs cleaned volatility
pre_std = df.groupby('Name')['Return'].std().mean()
post_std = df_iqr.groupby('Name')['Return'].std().mean()
print(f"Avg volatility before: {pre_std:.4f} | after cleaning: {post_std:.4f}")
```

```python
def SMA(df, n): 
    df[f'SMA_{n}'] = df['close'].rolling(n).mean()
def EMA(df, n): 
    df[f'EMA_{n}'] = df['close'].ewm(span=n, min_periods=n).mean()
def MOM(df, n): 
    df[f'MOM_{n}'] = df['close'].diff(n)
def ROC(df, n):
    df[f'ROC_{n}'] = df['close'].diff(n-1) / df['close'].shift(n-1)
def RSI(df, n):
    delta = df['close'].diff()
    gain = np.where(delta>0, delta, 0)
    loss = np.where(delta<0, -delta, 0)
    avg_gain = pd.Series(gain).rolling(n).mean()
    avg_loss = pd.Series(loss).rolling(n).mean()
    rs = avg_gain / avg_loss
    df[f'RSI_{n}'] = 100 - (100 / (1 + rs))
def BBANDS(df, n, multiplier=2):
    ma = df['close'].rolling(n).mean()
    sd = df['close'].rolling(n).std()
    df[f'BB_UP_{n}'] = ma + multiplier*sd
    df[f'BB_DOWN_{n}'] = ma - multiplier*sd
def ATR_true(df, n=14):
    prev_close = df['close'].shift(1)
    tr = pd.concat([
        df['high']-df['low'],
        (df['high']-prev_close).abs(),
        (df['low']-prev_close).abs()
    ], axis=1).max(axis=1)
    df[f'ATR_{n}'] = tr.ewm(span=n, min_periods=n).mean()
def WILLIAMS_R(df, n=14):
    hh = df['high'].rolling(n).max()
    ll = df['low'].rolling(n).min()
    df[f'WILLR_{n}'] = (hh - df['close']) / (hh - ll) * (-100)
def CCI_true(df, n=20):
    tp = (df['high'] + df['low'] + df['close']) / 3
    sma_tp = tp.rolling(n).mean()
    md = (tp - sma_tp).abs().rolling(n).mean()
    df[f'CCI_{n}'] = (tp - sma_tp) / (0.015 * md)
def CMF(df, n=20):
    mf_mult = ((df['close'] - df['low']) - (df['high'] - df['close'])) / (df['high'] - df['low'])
    mf_vol = mf_mult * df['volume']
    df[f'CMF_{n}'] = mf_vol.rolling(n).sum() / df['volume'].rolling(n).sum()
def AROON(df, n=25):
    aroon_up = df['high'].rolling(n+1).apply(lambda x: np.argmax(x[::-1]), raw=True)
    aroon_down = df['low'].rolling(n+1).apply(lambda x: np.argmax(x[::-1]), raw=True)
    df[f'AROON_UP_{n}'] = 100 * (n - aroon_up) / n
    df[f'AROON_DOWN_{n}'] = 100 * (n - aroon_down) / n
def MACD_hist(df, n_fast=12, n_slow=26, n_signal=9):
    ema_fast = df['close'].ewm(span=n_fast, min_periods=n_fast).mean()
    ema_slow = df['close'].ewm(span=n_slow, min_periods=n_slow).mean()
    macd = ema_fast - ema_slow
    signal = macd.ewm(span=n_signal, min_periods=n_signal).mean()
    df[f'MACD_HIST_{n_fast}_{n_slow}_{n_signal}'] = macd - signal
def ACCDIST(df, n):
    ad = (2 * df['close'] - df['high'] - df['low']) / \
        (df['high'] - df['low']) * df['volume']
    roc = ad.diff(n - 1) / ad.shift(n - 1)

    df['ACCDIST_' + str(n)] = roc
def CHAIKIN(df):
    ad = (2 * df['close'] - df['high'] - df['low']) / \
        (df['high'] - df['low']) * df['volume']
    Chaikin = pd.Series.ewm(ad, span=3, min_periods=2).mean(
    ) - pd.Series.ewm(ad, span=10, min_periods=9).mean()

    df['CHAIKIN'] = Chaikin
def OBV_true(df):
    direction = np.sign(df['close'].diff())
    df['OBV'] = (direction * df['volume']).cumsum()
def EOM(df, n):
    EoM = (df['high'].diff(1) + df['low'].diff(1)) * \
        (df['high'] - df['low']) / (2 * df['volume'])

    df['EOM_' + str(n)] = pd.Series.rolling(EoM, n).mean()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import numpy as np
import pandas as pd

def hybrid_scale_for_plot(df, cols):
    """
    Automatically choose between StandardScaler and MinMaxScaler 
    based on feature family or value range.
    """
    tmp = df[cols].replace([np.inf, -np.inf], np.nan).dropna()
    tmp["Name"] = df.Name.values
    bounded_keywords = ['RSI', 'AROON', 'CMF', 'WILLR']
    bounded_cols = [c for c in cols if any(k in c for k in bounded_keywords)]
    unbounded_cols = [c for c in cols if c not in bounded_cols]

    df_scaled = pd.DataFrame(index=tmp.index)
    
    # MinMax scale bounded features
    if bounded_cols:
        scaler_minmax = MinMaxScaler()
        scaled_bounded = scaler_minmax.fit_transform(tmp[bounded_cols])
        df_scaled[bounded_cols] = scaled_bounded

    # Standard scale unbounded features
    if unbounded_cols:
        scaler_standard = StandardScaler()
        scaled_unbounded = scaler_standard.fit_transform(tmp[unbounded_cols])
        df_scaled[unbounded_cols] = scaled_unbounded
    df_scaled["Name"] = tmp["Name"].values
    return df_scaled
```

```python
# ================================================
# 📊 MARKET-LEVEL DISTRIBUTIONS (ALL STOCKS)
# ================================================
for family, cols in feature_families.items():
    available_cols = [c for c in cols if c in df_feat_clean.columns]
    if not available_cols:
        continue
    # 🧠 Scale once per family
    tmp_scaled = hybrid_scale_for_plot(df_feat_clean, available_cols)
    combined = tmp_scaled[available_cols].values.flatten()
    lower, upper = np.nanpercentile(combined, [1, 99])  # <-- dynamic cutoff
    plt.figure(figsize=(10,6))
    for col in available_cols:
        sns.kdeplot(data=tmp_scaled, x=col, fill=True, alpha=0.25, label=col)
    plt.title(f"Market-Wide Distribution — {family} Features", fontsize=16)
    plt.xlabel("Normalized Feature Value")
    plt.xlim(lower,upper)
    plt.ylabel("Density")
    plt.legend(title=family, bbox_to_anchor=(1.05,1), loc='upper left')
    plt.tight_layout()
    plt.show()

```
